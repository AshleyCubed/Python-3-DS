```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import dataframe
df = pd.read_csv("movie_ratings.csv")
```

```python
#view column names
df.columns
```

```python
#rename columns
df.rename(columns={'Adjusted Gross ($mill)':'Adjusted_Gross', 
                   'Budget ($mill)':'Budget', 
                   'Gross ($mill)':'Gross', 
                   'Overseas ($mill)':'Overseas', 
                   'US ($mill)':'US', 
                   'Movie Title':'Movie', 
                   'Release Date':'Release_Date', 
                   'Runtime (min)':'Runtime'}, inplace=True)
```

```python
#view first 5 lines
df.head()
```

```python
#view summary stats of numeric variables
df.describe()
```

## Line Plot & Logarithmic Scale ##

```python
plt.plot(pd.to_datetime(pd.Series(df['Release_Date'])), df['Budget'])
#plt.yscale('log')
```

## Histogram ##

```python
plt.hist(df['Runtime'])
```

## Bar Charts ##

```python
df.Genre.value_counts().plot(kind = "bar")
```

```python
df.groupby('Genre').Budget.mean().plot(kind = "bar")
```

## Boxplot & X Axis Rotation for Readable Text##

```python
df.boxplot(column = 'US', by = 'Genre' )
plt.xticks(rotation=90)
```

## Customization ##

```python
df.groupby('Release_Date').Budget.mean().plot(kind = "line")
#plt.plot(pd.to_datetime(pd.Series(df['Release_Date'])), df['Budget'])
#plt.yscale('log')
plt.xlabel('Release Date')
plt.ylabel('Movie Budget')
plt.title('Movie Budgets by Release Date')
tick_val = [0,100, 200, 300]
tick_lab = ['0MM','100M', '200M', '300M'] 
plt.yticks(tick_val, tick_lab)
```
